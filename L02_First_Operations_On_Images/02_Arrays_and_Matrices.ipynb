```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.font_manager as font_manager

# define font for plot
font_legend = font_manager.FontProperties(family='DejaVu Sans',
                                   weight='normal',
                                   style='normal', size=20)
```

# Selecting arrays

## Selecting a part of an array
The following exercise shows how to:
- select a single element
- select a number of first elements
- select a number of last elements
- select a range of elements within an array
- sample an array (every n-th element)


```python
# here, we will define the array for further processing
a_val = 100*np.random.rand(20)

plt.figure(figsize=(20,5))
plt.plot(a_val)
```

```python
# we will examine the shape of the array
print('a_val has {0} dimension:s with {1} elements.'.format(a_val.ndim, a_val.size))
```

### Selecting a single element

```python
# indexing starts from 0, so, to select the first element, use a_val[0]
a_val[0]
```

```python
# last element is then the lenght of the array - 1, or just -1
print('last element is a_val["the lenght of the array" - 1] = {0},'\
      ' or just a_val[- 1] = {1}'.format(a_val[a_val.size-1], a_val[-1]))
```

### Selecting a part of an array

```python
# the following code is to define the x values in plots
x_vals = np.arange(1, a_val.size+1,1)

fig, axs = plt.subplots(4,1, figsize=(20, 20), sharex=True) # This sharex means that x_axis will be shared among plots
plt.xticks(x_vals)

# plot the whole array
axs[0].plot(x_vals, a_val, marker='o')
axs[0].grid()
axs[0].set_title('the whole array')
axs[0].tick_params(labelbottom=True)

# plot 7 first elements of array
axs[1].plot(x_vals[:7], a_val[0:7], marker='o') # counting starts with 0 and the last element is not included
axs[1].grid()
axs[1].set_title('the first 7 elements')
axs[1].tick_params(labelbottom=True)

# plot 7 last elements array
axs[2].plot(x_vals[-7:], a_val[-7:], marker='o')
axs[2].grid()
axs[2].set_title('the last 7 elements')
axs[2].tick_params(labelbottom=True)

# plot from the 11th to 17th element of the array
axs[3].plot(x_vals[10:16], a_val[10:16], marker='o')
axs[3].grid()
axs[3].set_title('from 11th to 17th element (notice that the last element is not included.)')
axs[3].tick_params(labelbottom=True)

# make sure that while plotting using values of both axes: plot(x, y), both axes are given an array of the same length.
```

### Sampling an array - selecting every n-th element

```python
# the following code is to define the x values in plots
x_vals = np.linspace(1, a_val.size, a_val.size) # this is another way to define the x axes #
                                                    # np.linspace(first_elem, last_elem, num_of_elements) 

#fig, axs = plt.subplots(3,1, figsize=(20, 20), sharex=True) # This sharex means that x_axis will be shared among plots

plt.figure(figsize=(20,10))
plt.xticks(x_vals)

# plot the whole array
plt.plot(x_vals, a_val, color='black', linestyle='dashed', marker='.', label='the whole array')

# plot every 3rd element
plt.plot(x_vals[::3], a_val[::3], marker='s', label='plot every 3rd element') # counting starts with 0 and the last element is not included

# plot every 2rd element from the 11th to 20th element
plt.plot(x_vals[10:20:2], a_val[10:20:2], marker='<', label='plot every 2nd element') # counting starts with 0 and the last element is not included
plt.grid()
plt.legend(prop=font_legend)
```

To plot a matrix, we can use:
- surface plot
- wireframe plot
- *image view*

We will first define a 2D-array to plot.

```python
mat_vals = np.zeros(shape=(20,20))
x_vals = np.linspace(-4*np.pi,4*np.pi,20)

# here, we are just generating some matrix data
mat_koef=1
mat_scale=1
for idx in range(mat_vals.shape[1]):
    mat_vals[:,idx] = mat_scale*np.sin(x_vals/mat_koef)
    mat_koef+=0.03
    mat_scale+=0.5
#-------------------------------------------------

# plotting    
fig = plt.figure(figsize=(15,10))

# this is needed to generate a grid for 3D representation
X, Y = np.meshgrid(x_vals, x_vals)

# surface plot
ax = fig.add_subplot(221, projection='3d')
ax.view_init(elev=40, azim=220)
ax.dist=9
p = ax.plot_surface(X, Y, mat_vals, cmap='plasma', alpha=0.70)
ax.set_title('Surface plot')
fig.colorbar(p)

# wireframe plot
ax = fig.add_subplot(222, projection='3d')
ax.view_init(elev=60, azim=300)
ax.dist=10
ax.plot_wireframe(X, Y, mat_vals)
ax.set_title('Wireframe plot')

# image representation
ax = fig.add_subplot(223)
ax.imshow(mat_vals)
ax.set_title('Image representation')

# contour representation
ax = fig.add_subplot(224)
ax.contour(mat_vals, cmap='YlGnBu_r', zorder=1)
ax.set_title('Contour represenation')
```
